# What is Grid?

OpenGrid (Grid for short) is a collaborative network platform of data owners and data scientists who can collectively train and share AI models using PySyft/Pytorch libraries.


## Overview
- [Introduction](#Introduction)
- [Objectives](#Objectives)
- [Implementation](#Implementation)

### Introduction

Grid platform aims to be a secure peer to peer platform used to train, manage and share models. A Grid is a collection of workers that can exchange messages and tensors in a transparent way.


Grid currently supports:

* Exchange of messages between remote workers like you usually would do using PySyft with Virtual Workers
* Machine Learning as a Service (MLaS): hosting and querying models using Grid
* Encrypted Machine Learning as a Service (EMLaS): hosting and querying encrypted models using Grid
* Support for deploying and running workers on Heroku

Grid will soon support:

* Federated Learning processes with no need to manage distributed workers directly.


[Try our tutorials to learn how to use Grid!](https://github.com/OpenMined/Grid/tree/dev/examples)


### Objectives

- ***Create a secure federated learning platform.***
  - Protect Intelectual Property
  - Protect data privacy
  - Mitigate misuse
- ***Create a collaborative federated learning platform.***
  - Allow users to send plain text / anonymized datasets that can be used by other users at the grid platform.
  - Allow users to send plain text / anonymized models that can be used/improved by other users at the grid platform.
  - Perform federated learning process using datasets available at the grid platform.
- ***Serve models in a secure way***
  - Run remote plain text / anonymized inferences on models hosted on grid network.
  - Download models hosted on grid network.
  - Perform arbitrary arithmetical / logic computing using anonymized datasets.
  - Provide fault tolerance.
  
  


### Implementation

The Grid repository consists mainly of 3 folders:

#### grid

`grid` is a python framework that extends PySyft with new workers that are capable of performing remote communication via a Rest API or via websockets. Grid also adds support for deploying these new clients on different plataforms (such as Heroku).


#### app

`apps` are basically regular Flask servers with a postgresql database. We add to these servers the ability to talk to Grid Workers, store tensors and models on the database and fault tolerance.

The idea is that Grid Workers are a way to communicate with an app.

#### grid_gateway

`grid_gateway` works like a special DNS component but it will route nodes by queries instead of domain names.
 The Grid Gateway needs to know the addresses and ids of all grid nodes connected on the Grid Network (workers that are all connected to each other).

A user can talk to a Grid Gateway and seemingless perform operations on multiple remote nodes without even knowing that they exist on the network.
