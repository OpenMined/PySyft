```python
for row in range(2, row_count + 1):
    
    # document type
    try:
        if curr_doc_type:= get(ws, 'DocType', row):
            put(ws, 'DocType', row, doctype[curr_doc_type])
    except:
        pass

    parked = False
    if get(ws, 'Status', row) == 'Parked':
        parked = True

    # payment block
    if get(ws, 'PaymentBlock', row):
        #TODO: add status check
        put(ws, 'Status', row, 'Blocked for ***')

    # mapping PC codes to comments/actions
    if text_str:= get(ws, 'Text', row):
        if pc_search:=re.search(r'(PC)\d\d',str(text_str)):
            pc_action = pc_actions[pc_search[0].upper()]
            pc_comment = pc_comments[pc_search[0].upper()]
            put(ws, 'Comment', row, pc_comment)
            put(ws, 'ActionRequired', row, pc_action)

    # getting scheduled date for non-parked items
    if not parked:
        try:
            if baseline_date:= get(ws, 'BaselineDate', row):
#                 baseline_date = datetime.strptime(baseline_date, '%d.%m.%Y')
                scheduled_date = get_scheduled_date(baseline_date, vendor_type)
                put(ws, 'ScheduledDate', row, scheduled_date)

                # update total balance for scheduled date
                amount_sgn = get(ws, 'Credit/Debit', row)
                amount = get(ws, 'LCAmount', row)
                
                if amount_sgn == 'H':
                    amount *= -1
#                 amount = get(ws, 'LCAmount', row).split(',')
#                 if len(amount) > 1:
#                     amount = float(amount[0].replace('.', '') + '.' + amount[1])
#                 else:
#                     amount = float(amount[0].replace('.', ''))
                balance[scheduled_date] = balance.get(scheduled_date, 0) + amount
        except Exception as e:
                print(e)
                put(ws, 'ScheduledDate', row, '')
    else:
        put(ws, 'ScheduledDate', row, '')

# debit-balance check; if debit balance, update status and scheduled date
for row in range(2, row_count + 1):
    if balance.get(get(ws, 'ScheduledDate', row), -1) >= 0:
        put(ws, 'Status', row, 'Debit balance')
        put(ws, 'ScheduledDate', row, '')

# rename columns
for col in col_names:
    put(ws, col, 1, col)

# delete unnecessary columns
# to_delete = ['ReferenceKey', 'BaselineDate', 'PCAmount', 'Text','PaymentBlock']
to_delete = ['Status_V', 'Text', 'PaymentBlock']

for col in to_delete:
    ws.delete_cols(columns[col], 1)

try:
    workbook.save(filename = save_file_name)
except:
    logger.add_log('Unable to save modified VendorLine file.')
```

```python
import email
from openpyxl import load_workbook
import smtplib,ssl
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email.utils import formatdate
from email import encoders
from datetime import datetime
import re
import os
import sys
import numpy
import smtplib
import keyring
import pandas as pd


use_cmd_args = False
folder_location = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29"
fbl1n_file_name = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29\FBL1N_381111.xlsx"
vim_file_name = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29\VIM_Analytics_381111.xlsx"
save_file_name = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29\Status-381111-" + str(datetime.now().strftime("%Y-%m-%d-%H-%M-%S")) + ".xlsx"
log_file_name = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29\log_381111_405.txt"
vendor_id = 381111
vendor_type = 1
vendor_name = "ADM Spain"
pc_codes_file = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29\pc_codes.xlsx"
euz_scheduled_run_file = r"C:\Users\40103611\Desktop\PMI_EUR_BOT_29\EUZ_Scheduled_Run.xlsx"
vendor_email_address = 'vaibhav.vardhan@ab-inbev.com'

folder_location = '/Users/vsquareg/AbInBev'
fbl1n_file_name = os.path.join(folder_location, '0000381111_raw.xlsx')
save_file_name = os.path.join(folder_location, '0000381111_final.xlsx')
log_file_name = os.path.join(folder_location, '0000381111_log.txt')
vendor_id = '0000381111'
vendor_type = 4
pc_codes_file = os.path.join(folder_location, 'pc_codes.xlsx')
euz_scheduled_run_file = os.path.join(folder_location, 'EUZ_Scheduled_Run.xlsx')
vim_file_name = os.path.join(folder_location, 'VIM_Analytics_381111.xlsx')
vendor_name = 'ADM Spain'
vendor_email_address = os.path.join(folder_location, 'vaibhav.vardhan@ab-inbev.com')

if __name__ == "__main__" and use_cmd_args:
    folder_location = sys.argv[1]
    fbl1n_file_name = os.path.join(folder_location, sys.argv[2])
    save_file_name = os.path.join(folder_location, sys.argv[3])
    log_file_name = os.path.join(folder_location, sys.argv[4])
    vendor_id = sys.argv[5]
    vendor_type = sys.argv[6]
    pc_codes_file = os.path.join(folder_location, sys.argv[7])
    euz_scheduled_run_file = os.path.join(folder_location, sys.argv[8])
    vim_file_name = os.path.join(folder_location, sys.argv[9])
    vendor_email_address = sys.argv[10]
    vendor_name = sys.argv[11]


class Email:
    def __init__(self):
        self.client = smtplib.SMTP('smtp.gmail.com', 587)
  
        # start TLS for security
        self.client.starttls()

        # Authentication
        self.client.login("incupton@gmail.com", keyring.get_password('test_email', 'inc.upton@gmail.com'))

        self.from_email = "inc.upton@gmail.com"
        
    def send_email(self, to, subject, body, attachment = None):
        msg = MIMEMultipart()
        msg['From'] = self.from_email
        msg['To'] = to
        msg['Date'] = formatdate(localtime = True)
        msg['Subject'] = subject
        msg.attach(MIMEText(body))

        if attachment:
            fp = open(attachment, 'rb')
            part = MIMEBase('application','vnd.ms-excel')
            part.set_payload(fp.read())
            fp.close()
            encoders.encode_base64(part)
            part.add_header('Content-Disposition', 'attachment', filename=os.path.basename(attachment))
            msg.attach(part)

        self.client.sendmail(self.from_email, to, msg.as_string())

    def __del__(self):
        try:
            self.client.quit()
        except:
            pass
    
        
class Logger:
    def __init__(self, log_file_name, vendor_id):
        self.file = open(log_file_name, "a")
        self.vendor_id = vendor_id

    def add_log(self, err):
        self.file.write(' [ Execution for vendor ID: ' + str(self.vendor_id) + ' failed. ' + err + '] ')

    def __del__(self):
        self.file.close()

def process_ref_key(tmp):
    if not numpy.isnan(tmp):
        return int(tmp/10000)
    return tmp

def get(ws, col_name, row, value = True):
    if value:
        return ws.cell(row = row, column = columns[col_name]).value
    return ws.cell(row = row, column = columns[col_name])

def put(ws, col_name, row, val):
    ws.cell(row = row, column = columns[col_name]).value = val

def extract_PC_codes(pc_codes_file):
   workbook = load_workbook(filename=pc_codes_file)
   ws = workbook[workbook.sheetnames[0]]
   for row in ws.iter_rows(min_row = 2):
       pc_actions[row[0].value] = row[3].value
       pc_comments[row[0].value] = row[1].value

def handle_datetime(val):
    if 'datetime' not in str(type(val)):
        val = datetime.strptime(val, '%d/%m/%Y')
    return val
    
def extract_scheduled_dates(euz_scheduled_run_file):
    workbook = load_workbook(filename=euz_scheduled_run_file)
    ws = workbook[workbook.sheetnames[0]]
    for row in ws.iter_rows(min_row = 2):
        payment_dates.append(handle_datetime(row[0].value))
        cutoff_dates.append(handle_datetime(row[1].value))

def get_scheduled_date(baseline_date, vendor_type = 1):
    row_num = 12
    lowest_delta = 357
    scheduled_date = payment_dates[0]
    if vendor_type == 1:
        row_num = 12
    elif vendor_type == 2:
        row_num = 24
    elif vendor_type == 4:
        row_num = 48
    
    for row in range(row_num):
        now_delta = (payment_dates[row] - datetime.now()).days
        posting_delta = (cutoff_dates[row] - baseline_date).days
        # print(payment_dates[row], baseline_date, now_delta, posting_delta)
        if now_delta >= 0 and now_delta < lowest_delta and posting_delta >= 0:
            scheduled_date = payment_dates[row]
            lowest_delta = now_delta
    
    return scheduled_date.date()

def process_ip_number(x):
    if numpy.isnan(x['Purch.Doc.']):
        return float('nan')
    if numpy.isnan(x['Old DocNum']):
        return x['DocumentNo_y']
    else:
        return x['Old DocNum']

col_names = ['VendorID', 'PaymentBlock', 'DocType', 'CompanyCode', 'DocumentNo', 'Reference', 'PaymentTerms', 'Credit/Debit', 'LCAmount', 'Currency', 'TXCode', 'Text', 'BaselineDate', 'Status_V', 'Status', 'ReferenceKey', 'Comment', 'ActionRequired', 'ScheduledDate']

doctype = {
    'YE': 'Basware type',
    'RE': 'B2B/PDF',
    'KG': 'FI Credit Note',
    'KR': 'Credit Note',
    }

columns = {}
pc_actions = {}
pc_comments = {}
payment_dates = []
cutoff_dates = []

balance = {}


logger = Logger(log_file_name, vendor_id)
email_handler = Email()

import time
start = time.time()

for i, col in enumerate(col_names):
    columns[col] = i+1

try:
    extract_PC_codes(pc_codes_file)
except:
    logger.add_log('Unable to extract PC codes.')
    raise

try:
    extract_scheduled_dates(euz_scheduled_run_file)
except:
    logger.add_log('Unable to extract EUZ_Scheduled_Run')
    raise

begin_row = 1
begin_col = 1

try:
    workbook = load_workbook(filename=fbl1n_file_name)
    ws = workbook[workbook.sheetnames[0]]
except:
    logger.add_log('Unable to load VendorLine file.')
    raise

row_count = ws.max_row
# row_count = 40

for row in range(2, row_count + 1):
    
    # document type
    try:
        if curr_doc_type:= get(ws, 'DocType', row):
            put(ws, 'DocType', row, doctype[curr_doc_type])
    except:
        pass

    parked = False
    if get(ws, 'Status', row) == 'Parked':
        parked = True

    # payment block
    if get(ws, 'PaymentBlock', row):
        #TODO: add status check
        put(ws, 'Status', row, 'Blocked for ***')

    # mapping PC codes to comments/actions
    if text_str:= get(ws, 'Text', row):
        if pc_search:=re.search(r'(PC)\d\d',str(text_str)):
            pc_action = pc_actions[pc_search[0].upper()]
            pc_comment = pc_comments[pc_search[0].upper()]
            put(ws, 'Comment', row, pc_comment)
            put(ws, 'ActionRequired', row, pc_action)

    # getting scheduled date for non-parked items
    if not parked:
        try:
            if baseline_date:= get(ws, 'BaselineDate', row):
#                 baseline_date = datetime.strptime(baseline_date, '%d.%m.%Y')
                scheduled_date = get_scheduled_date(baseline_date, vendor_type)
                put(ws, 'ScheduledDate', row, scheduled_date)

                # update total balance for scheduled date
                amount_sgn = get(ws, 'Credit/Debit', row)
                amount = get(ws, 'LCAmount', row)
                
                if amount_sgn == 'H':
                    amount *= -1
#                 amount = get(ws, 'LCAmount', row).split(',')
#                 if len(amount) > 1:
#                     amount = float(amount[0].replace('.', '') + '.' + amount[1])
#                 else:
#                     amount = float(amount[0].replace('.', ''))
                balance[scheduled_date] = balance.get(scheduled_date, 0) + amount
        except Exception as e:
                print(e)
                put(ws, 'ScheduledDate', row, '')
    else:
        put(ws, 'ScheduledDate', row, '')

# debit-balance check; if debit balance, update status and scheduled date
for row in range(2, row_count + 1):
    if balance.get(get(ws, 'ScheduledDate', row), -1) >= 0:
        put(ws, 'Status', row, 'Debit balance')
        put(ws, 'ScheduledDate', row, '')

# rename columns
for col in col_names:
    put(ws, col, 1, col)

# delete unnecessary columns
# to_delete = ['ReferenceKey', 'BaselineDate', 'PCAmount', 'Text','PaymentBlock']
to_delete = ['Status_V', 'Text', 'PaymentBlock']

for col in to_delete:
    ws.delete_cols(columns[col], 1)

try:
    workbook.save(filename = save_file_name)
except:
    logger.add_log('Unable to save modified VendorLine file.')
```
