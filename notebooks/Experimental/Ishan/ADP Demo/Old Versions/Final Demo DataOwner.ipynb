```python
import hashlib
from math import isnan, nan

hash_db = {}
hash_db[nan] = nan

def convert_string(s: str, digits: int=15):
    """Maps a string to a unique hash using SHA, converts it to a hash or an int"""
    if type(s) is str:
        new_hash = int(hashlib.sha256(s.encode("utf-8")).hexdigest(), 16) % 10**digits
        hash_db[s] = new_hash
        return new_hash
    else:
        return s
```

```python
duplicate_dataset = canada_dataset
string_cols = ['Classification', 'Period Desc.', 'Trade Flow','Reporter', 'Partner','Commodity']

for col in canada_dataset.columns:
    print(col)
    canada_dataset[col] = canada_dataset[col].map(lambda x: convert_string(x))
```
