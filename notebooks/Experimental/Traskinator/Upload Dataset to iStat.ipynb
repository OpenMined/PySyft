```python
# slice data we want to upload for analysis
data_batch = ((np.array(list(data['Trade Value (US$)'])) / 10000)[0:10]).astype(np.int32)

# slice references to real-world entities whose privacy we want to protect
trade_partners = ((list(data['Partner'])))[0:10]
```

```python
# create "Entity" objects so that the Domain node's Privacy Accountant will know which 
# data subjects occur in which parts of the dataset
entities = list()
for i in range(len(trade_partners)):
    entities.append(Entity(name=trade_partners[i]))
```

```python
# create dataset
sampled_italy_dataset = sy.Tensor(data_batch)

# annotate public shape (this will go away in a version soon but for now we have to manually annotate this)
sampled_italy_dataset.public_shape = sampled_italy_dataset.shape

# create privacy annotated dataset
sampled_italy_dataset = sampled_italy_dataset.private(0, 3, entity=entities[0]).tag("trade_flow")
```

```python
istat.load_dataset(
    assets={"feb2020-10": sampled_italy_dataset},
    name="Italy Trade Data - First few rows",
    description="""A collection of reports from iStat's statistics 
                    bureau about how much it thinks it imports and exports from other countries.""",
)
```
