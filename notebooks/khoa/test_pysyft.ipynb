```python
netherlands_trade_data_url = "https://github.com/OpenMined/PySyft/blob/dev/notebooks/trade_demo/datasets/nl%20-%20feb%202021.csv?raw=true"
data = pd.read_csv(netherlands_trade_data_url)
```

```python
# Filter commodities imported to Netherland
commodities_imported_to_netherlands = data[data['Trade Flow'] == "Imports"]
commodities_imported_to_netherlands
```

```python
# Trade Value of the commodities imported to Netherlands
trade_value_of_imported_commodites = commodities_imported_to_netherlands['Trade Value (US$)'].to_numpy()
```

```python
trade_value_of_imported_commodites
```

```python
# Countries from which commodities are imported to Netherlands
importers = commodities_imported_to_netherlands["Partner"]
importers
```

```python
data_subjects = sy.DataSubjectArray.from_objs(importers)
data_subjects
```

```python
min_val = trade_value_of_imported_commodites.min()
max_val = trade_value_of_imported_commodites.max()
```

```python
data = sy.Tensor(trade_value_of_imported_commodites).annotated_with_dp_metadata(
   min_val=float(min_val), max_val=float(max_val), data_subjects=data_subjects
)
```
