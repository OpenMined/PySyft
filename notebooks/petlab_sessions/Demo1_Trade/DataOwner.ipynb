### Annotate Data

```python
imports = canada_dataset[canada_dataset['Trade Flow'] == "Imports"]
exports = canada_dataset[canada_dataset['Trade Flow'] == "Exports"]
```

```python
imports_value = imports['Trade Value (US$)'].to_numpy()
exports_value = exports['Trade Value (US$)'].to_numpy()
```

```python
importers = imports["Partner"]
exporters = exports["Partner"]
```

```python
import syft as sy
```

```python
imports_data = sy.Tensor(child=imports_value).private(min_val=0, max_val=2e12, data_subjects=importers)
exports_data = sy.Tensor(child=exports_value).private(min_val=0, max_val=2e12, data_subjects=exporters)
```
