## Alice Python Server

```python
# third party
import requests
```

### 1. Create DHT Key and Private Route

```python
host = "localhost"
port = 4000
```

```python
res = requests.post(f"http://{host}:{port}/generate_dht_key")
```

```python
res.content
```

```python
res = requests.get(f"http://{host}:{port}/retrieve_dht_key")
```

```python
self_dht_key = res.json()["message"]
print("=" * 30)
print(self_dht_key)
print("=" * 30)
```

### Send AppMessage using DHT Key to Self

```python
json_data = {"dht_key": self_dht_key, "message": "Hello to me again"}
app_message = requests.post(f"http://{host}:{port}/app_message", json=json_data)
```

```python
app_message.content
```

### Send AppCall using DHT Key to Self

```python
json_data = {"dht_key": self_dht_key, "message": "Hello to app call"}
app_call = requests.post(f"http://{host}:{port}/app_call", json=json_data)
```

```python
app_call.json()
```

### Send AppMessage using DHT Key to Peer

```python
peer_dht_key = input("Enter Peer DHT Key")
```

```python
json_data = {"dht_key": peer_dht_key, "message": "How are you doing , Bob"}
app_message = requests.post(f"http://{host}:{port}/app_message", json=json_data)
```

### Send Proxy Message 

```python
res = requests.get(
    f"http://{host}:{port}/proxy",
    json={"url": "https://www.google.com", "dht_key": self_dht_key},
)
```

```python
res.content
```
